### Visualization
Straigh from Scipy docs: https://docs.scipy.org/doc/scipy/reference/generated/scipy.io.wavfile.read.html

```python
samplerate, data = sio.wavfile.read(wav_fname)
print(f"number of channels = {data.shape[1]}")

length = data.shape[0] / samplerate
print(f"length = {length}s")

print(f"array length = {data.shape[0]}")
```

#### Basic Waveform Plot

```python
import matplotlib.pyplot as plt
import numpy as np

chunk = data.shape[0]/10 # take 1 second chunk

time = np.linspace(0., length, data.shape[0])
plt.plot(time[:100], data[:100, 0], label="Left channel") # plotting first 100 spots
plt.plot(time[:100], data[:100, 1], label="Right channel")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
